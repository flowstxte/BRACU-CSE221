```python
# Open input and output files
file_inp = open("input1.txt", "r")
file_out = open("output1.txt", "w")

# Read the number of nodes, edges, and build the graph
num_nodes, num_edges = map(int, file_inp.readline().split())
graph = [[0] * num_nodes for _ in range(num_nodes)]
for _ in range(num_edges):
    u, v, w = map(int, file_inp.readline().split())
    graph[u - 1][v - 1] = w

# Read the source node
source = int(file_inp.readline()) - 1

# Dijkstra's algorithm
def dijkstra(graph, source):
    distances = [float("inf")] * len(graph)
    distances[source] = 0
    visited = [False] * len(graph)

    for _ in range(len(graph)):
        # Find the unvisited node with the smallest distance
        min_node = -1
        min_dist = float("inf")
        for i in range(len(graph)):
            if not visited[i] and distances[i] < min_dist:
                min_dist = distances[i]
                min_node = i

        if min_node == -1:  # No more reachable nodes
            break

        visited[min_node] = True

        # Update distances to neighbors
        for neighbor in range(len(graph)):
            if graph[min_node][neighbor] > 0 and not visited[neighbor]:
                new_dist = distances[min_node] + graph[min_node][neighbor]
                if new_dist < distances[neighbor]:
                    distances[neighbor] = new_dist

    # Replace `float("inf")` with -1 for unreachable nodes
    for i in range(len(distances)):
        if distances[i] == float("inf"):
            distances[i] = -1

    return distances

# Find shortest paths and write to the output file
distances = dijkstra(graph, source)
file_out.write(" ".join(map(str, distances)) + "\n")

# Close the files
file_inp.close()
file_out.close()
```

```python
file_inp = open("input2.txt", "r")
file_out = open("output2.txt", "w")

n, m = map(int, file_inp.readline().split())
edges = []
for _ in range(m):
    u, v, w = map(int, file_inp.readline().split())
    edges.append((u, v, w))
s, t = map(int, file_inp.readline().split())

import heapq as hq

def dijkstra(adj_list, start, n):
    distances = [float('inf')] * (n + 1)
    distances[start] = 0
    pq = [(0, start)]
    while pq:
        current_distance, current_node = hq.heappop(pq)
        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in adj_list[current_node]:
            new_distance = current_distance + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                hq.heappush(pq, (new_distance, neighbor))
    return distances

def find_meeting_node(n, m, edges, s, t):
    adj_list = [[] for _ in range(n + 1)]
    for u, v, w in edges:
        adj_list[u].append((v, w))
    dist_from_s = dijkstra(adj_list, s, n)
    dist_from_t = dijkstra(adj_list, t, n)

    min_time, meeting_node = float('inf'), -1
    for node in range(1, n + 1):
        if dist_from_s[node] != float('inf') and dist_from_t[node] != float('inf'):
            max_time = max(dist_from_s[node], dist_from_t[node])
            if max_time < min_time:
                min_time, meeting_node = max_time, node
    return "Impossible" if meeting_node == -1 else f"Time {min_time}\nNode {meeting_node}"

result = find_meeting_node(n, m, edges, s, t)
file_out.write(result)

file_inp.close()
file_out.close()
```

```python
file_inp = open("input3.txt", "r")
file_out = open("output3.txt", "w")

import heapq as hq
num_nodes, num_edges = map(int, file_inp.readline().split())
graph = {i: [] for i in range(num_nodes + 1)}
for _ in range(num_edges):
    node_from, node_to, cost = map(int, file_inp.readline().split())
    graph[node_from].append((cost, node_to))

def dijkstra(graph, start, end):
    queue = [(0, start)]
    visited = [False] * len(graph)
    parent = [None] * len(graph)
    distance = [float('inf')] * len(graph)
    distance[start] = 0
    while queue:
        dist, node = hq.heappop(queue)
        if not visited[node]:
            visited[node] = True
            for danger, neighbor in graph[node]:
                new_danger = max(distance[node], danger)
                if new_danger < distance[neighbor]:
                    distance[neighbor] = new_danger
                    parent[neighbor] = node
                    hq.heappush(queue, (distance[neighbor], neighbor))
    while float('inf') in distance:
        idx = distance.index(float('inf'))
        distance[idx] = -1

    return distance[end]
source_node = 1
file_out.write(f"{dijkstra(graph, source_node, num_nodes)}")

file_inp.close()
file_out.close()
```

```python
file_inp = open("input4.txt", "r")
file_out = open("output4.txt", "w")

n, k = map(int, file_inp.readline().split())
queries = [tuple(map(int, file_inp.readline().split())) for _ in range(k)]

class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n + 1))
        self.size = [1] * (n + 1)

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        rootX = self.find(x)
        rootY = self.find(y)

        if rootX != rootY:
            if self.size[rootX] < self.size[rootY]:
                rootX, rootY = rootY, rootX
            self.parent[rootY] = rootX
            self.size[rootX] += self.size[rootY]

    def get_size(self, x):
        root = self.find(x)
        return self.size[root]

def process_queries(n, k, queries):
    uf = UnionFind(n)
    result = []
    for a, b in queries:
        uf.union(a, b)
        result.append(str(uf.get_size(a)))

    return "\n".join(result)

result = process_queries(n, k, queries)
file_out.write(result)

file_inp.close()
file_out.close()
```

```python
file_inp = open("input5.txt", "r")
file_out = open("output5.txt", "w")

n, m = map(int, file_inp.readline().split())
edges = [tuple(map(int, file_inp.readline().split())) for _ in range(m)]

def find(parent, x):
    if parent[x] != x:
        parent[x] = find(parent, parent[x])
    return parent[x]

def union(parent, size, x, y):
    rootX = find(parent, x)
    rootY = find(parent, y)

    if rootX != rootY:
        if size[rootX] < size[rootY]:
            rootX, rootY = rootY, rootX
        parent[rootY] = rootX
        size[rootX] += size[rootY]

def kruskal(n, edges):
    parent = list(range(n + 1))
    size = [1] * (n + 1)
    edges.sort(key=lambda edge: edge[2])
    mst_cost = 0
    edges_used = 0

    for u, v, w in edges:
        if find(parent, u) != find(parent, v):
            union(parent, size, u, v)
            mst_cost += w
            edges_used += 1
            if edges_used == n - 1:
                break
    return mst_cost

result = kruskal(n, edges)
file_out.write(str(result))

file_inp.close()
file_out.close()
```
