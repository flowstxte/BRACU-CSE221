```python
# Open input and output files
file_inp = open("input2.txt", "r")
file_out = open("output2.txt", "w")
# Read the number of vertices and edges
ver, edg = map(int, file_inp.readline().split())


# Initialize adjacency dist for the graph
adj = {i: [] for i in range(ver + 1)}

# Build the graph
for _ in range(edg):
    u, v = map(int, file_inp.readline().split())
    adj[u].append(v)
    adj[v].append(u)

# Breadth-First Search (BFS) function
def bfs(graph, start):
    visited = [False] * (len(graph))
    queue = [start]
    visited[start] = True
    result = []

    while queue:
        node = queue.pop(0)
        result.append(str(node))

        # Explore neighbors in sorted order
        for neighbor in sorted(graph[node]):
            if not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)

    return " ".join(result)

# Perform BFS starting from vertex 1 and write the result to the output file
file_out.write(bfs(adj, 1))

# Close the files
file_inp.close()
file_out.close()
```

```python
# Open input and output files
file_inp = open("input3.txt", "r")
file_out = open("output3.txt", "w")

# Read the number of vertices and edges
ver, edg = map(int, file_inp.readline().split())

# Initialize adjacency list for the graph
adj = {i: [] for i in range(int(ver) + 1)}

# Build the graph
for _ in range(int(edg)):
    u, v = map(int, file_inp.readline().split())
    adj[u].append(v)
    adj[v].append(u)

# Initialize visited list
visited = [False] * (int(ver) + 1)

# Depth-First Search (DFS) function
def dfs(graph, node):
    visited[node] = True
    result = [str(node)]

    for neighbor in sorted(graph[node]):  # Sort neighbors for consistent traversal
        if not visited[neighbor]:
            result.append(dfs(graph, neighbor))

    return " ".join(result)

# Perform DFS starting from vertex 1 and write the result to the output file
file_out.write(dfs(adj, 1))

# Close the files
file_inp.close()
file_out.close()
```

```python
# Open input and output files
file_inp = open("input5.txt", "r")
file_out = open("output5.txt", "w")

# Read the number of vertices, edges, and the destination vertex
ver, edg, end = map(int, file_inp.readline().split())

# Initialize adjacency list for the graph
adj = {i: [] for i in range(ver + 1)}

# Build the graph
for _ in range(edg):
    u, v = map(int, file_inp.readline().split())
    adj[u].append(v)
    adj[v].append(u)

# BFS to find the shortest path
def shortest_path(graph, start, end):
    visited = [False] * (ver + 1)
    parent = [None] * (ver + 1)
    distance = [float("inf")] * (ver + 1)

    # BFS initialization
    queue = [start]
    visited[start] = True
    distance[start] = 0

    while queue:
        current = queue.pop(0)
        for neighbor in graph[current]:
            if not visited[neighbor]:
                visited[neighbor] = True
                parent[neighbor] = current
                distance[neighbor] = distance[current] + 1
                queue.append(neighbor)
                if neighbor == end:
                    break  # Stop early if we reach the destination

    # Reconstruct the path from end to start
    path = []
    while end is not None:
        path.append(end)
        end = parent[end]
    return distance[path[0]], path[::-1]  # Reverse the path for correct order

# Find the shortest path from vertex 1 to the given destination
time, path = shortest_path(adj, 1, end)

# Write results to the output file
file_out.write(f"Time: {time}\nShortest Path: {' '.join(map(str, path))}\n")

# Close the files
file_inp.close()
file_out.close()
```

```python
file_inp=open("input6.txt", "r")
file_out=open("output6.txt", "w")

s = file_inp.readline().split()
nds = int(s[0])
edg = int(s[1])

grid = [list(file_inp.readline().strip()) for _ in range(nds)]

def dfs(grid, row, col, visited):
    if row < 0 or col < 0 or row >= len(grid) or col >= len(grid[0]) or grid[row][col] == '#' or visited[row][col]:
        return 0
    visited[row][col] = True
    d = 1 if grid[row][col] == 'D' else 0
    for dr, dc in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
        d += dfs(grid, row + dr, col + dc, visited)
    return d

def max_diamonds(grid):
    rows = len(grid)
    cols = len(grid[0])
    visited = [[False] * cols for _ in range(rows)]

    md = 0
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] != '#' and not visited[i][j]:
                cd = dfs(grid, i, j, visited)
                md = max(md, cd)
    return md

result = max_diamonds(grid)
file_out.write(f"{result}\n")

file_inp.close()
file_out.close()
```

```python
file_inp=open("input7.txt","r")
file_out=open("output7.txt","w")

n = int(file_inp.readline())
edg = []
for _ in range(n - 1):
    u, v = file_inp.readline().split()
    edg.append((int(u), int(v)))

def bfs(start, n, adj):
    visited = [-1] * (n + 1)
    queue = [start]
    visited[start] = 0
    idx = 0
    fn = start
    md = 0

    while idx < len(queue):
        node = queue[idx]
        idx += 1
        for j in adj[node]:
            if visited[j] == -1:
                visited[j] = visited[node] + 1
                queue.append(j)
                if visited[j] > md:
                    md = visited[j]
                    fn = j

    return fn, md

def find_cities(n, edges):
    adj = {i: [] for i in range(1, n + 1)}
    for u, v in edges:
        adj[u].append(v)
        adj[v].append(u)

    sn = 1
    fn, _ = bfs(sn, n, adj)
    en, _ = bfs(fn, n, adj)

    return fn, en

ca, cb = find_cities(n, edg)

file_out.write(f"{ca} {cb}\n")

file_inp.close()
file_out.close()
```

```python
file_inp=open("input8.txt","r")
file_out=open("output8.txt","w")

def bfs(start, graph, visited, colors):
    queue = [start]
    visited[start] = True
    colors[start] = 0
    c = [0, 0]
    c[0] = 1

    while queue:
        n = queue.pop(0)
        cc = colors[n]
        for j in graph[n]:
            if j not in visited:
                visited[j] = True
                colors[j] = 1 - cc
                c[colors[j]] += 1
                queue.append(j)
            elif colors[j] == cc:
                return 0
    return max(c)

def solve_case(n, fights):
    graph = {}
    for u, v in fights:
        if u not in graph:
            graph[u] = []
        if v not in graph:
            graph[v] = []
        graph[u].append(v)
        graph[v].append(u)
    visited = {}
    colors = {}
    m = 0
    for node in graph:
        if node not in visited:
            m += bfs(node, graph, visited, colors)
    return m

t = int(file_inp.readline().strip())
for i in range(1, t + 1):
    n = int(file_inp.readline().strip())
    fights = []
    for _ in range(n):
        u, v = file_inp.readline().split()
        fights.append((int(u), int(v)))
    result = solve_case(n, fights)
    file_out.write(f"Case {i}: {result}\n")

file_inp.close()
file_out.close()
```

```python
file_inp = open("input9a.txt", "r")
file_out = open("output9a.txt", "w")

N, M = map(int, file_inp.readline().split())
graph = {}

for _ in range(M):
    A, B = map(int, file_inp.readline().split())
    if A not in graph:
        graph[A] = []
    graph[A].append(B)

for i in range(1, N + 1):
    if i not in graph:
        graph[i] = []

def enqueue(queue, item):
    queue.append(item)

def dequeue(queue):
    return queue.pop(0)

def bfs_topological_sort(graph, N):
    in_degree = [0] * (N + 1)
    for node in graph:
        for neighbor in graph[node]:
            in_degree[neighbor] += 1

    queue = []
    result = ""

    for i in range(1, N + 1):
        if in_degree[i] == 0:
            enqueue(queue, i)

    while queue:
        node = dequeue(queue)
        result += f"{node} "
        if node in graph:
            for neighbor in sorted(graph[node]):
                in_degree[neighbor] -= 1
                if in_degree[neighbor] == 0:
                    enqueue(queue, neighbor)

    if len(result.split()) == N:
        return result.strip()
    else:
        return "IMPOSSIBLE"

bfs_result = bfs_topological_sort(graph, N)
file_out.write(bfs_result)

file_inp.close()
file_out.close()
```

```python
file_inp = open("input9b.txt", "r")
file_out = open("output9b.txt", "w")

N, M = map(int, file_inp.readline().split())
graph = {i: [] for i in range(1, N + 1)}

for _ in range(M):
    A, B = map(int, file_inp.readline().split())
    graph[A].append(B)

def dfs_topological_sort(graph, N):
    visited = [0] * (N + 1)
    result = []

    def dfs(node):
        if visited[node] == 1:
            return False
        if visited[node] == 0:
            visited[node] = 1
            for neighbor in sorted(graph[node]):
                if not dfs(neighbor):
                    return False
            visited[node] = 2
            result.append(node)
        return True

    for i in range(1, N + 1):
        if visited[i] == 0:
            if not dfs(i):
                return "IMPOSSIBLE"

    output = ""
    for i in range(len(result) - 1, -1, -1):
        output += str(result[i])
        if i > 0:
            output += " "
    return output

file_out.write(dfs_topological_sort(graph, N))

file_inp.close()
file_out.close()
```

```python
file_inp = open("input10.txt", "r")
file_out = open("output10.txt", "w")

N, M = map(int, file_inp.readline().split())
graph = {i: [] for i in range(1, N + 1)}
in_degree = {i: 0 for i in range(1, N + 1)}

for _ in range(M):
    A, B = map(int, file_inp.readline().split())
    graph[A].append(B)
    in_degree[B] += 1

def bfs_topological_sort(N):
    queue = []
    result = []
    for i in range(1, N + 1):
        if in_degree[i] == 0:
            queue.append(i)
    queue.sort()

    while queue:
        node = queue.pop(0)
        result.append(node)
        for neighbor in sorted(graph[node]):
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)
                queue.sort()

    if len(result) == N:
        return " ".join(map(str, result))
    return "IMPOSSIBLE"

result = bfs_topological_sort(N)
file_out.write(result)

file_inp.close()
file_out.close()
```

```python
file_inp = open("input11.txt", "r")
file_out = open("output11.txt", "w")

N, M = map(int, file_inp.readline().split())
graph = {i: [] for i in range(1, N + 1)}
reverse_graph = {i: [] for i in range(1, N + 1)}

for _ in range(M):
    u, v = map(int, file_inp.readline().split())
    graph[u].append(v)
    reverse_graph[v].append(u)

def dfs(graph, node, visited, stack):
    visited[node] = True
    for neighbor in graph[node]:
        if not visited[neighbor]:
            dfs(graph, neighbor, visited, stack)
    stack.append(node)

def reverse_dfs(graph, node, visited, component):
    visited[node] = True
    component.append(node)
    for neighbor in graph[node]:
        if not visited[neighbor]:
            reverse_dfs(graph, neighbor, visited, component)

def kosaraju_scc(N):
    visited = [False] * (N + 1)
    stack = []
    for node in range(1, N + 1):
        if not visited[node]:
            dfs(graph, node, visited, stack)

    visited = [False] * (N + 1)
    sccs = []

    while stack:
        node = stack.pop()
        if not visited[node]:
            component = []
            reverse_dfs(reverse_graph, node, visited, component)
            sccs.append(sorted(component))

    return sccs

sccs = kosaraju_scc(N)

for scc in sccs:
    file_out.write(" ".join(map(str, scc)) + "\n")

file_inp.close()
file_out.close()
```
